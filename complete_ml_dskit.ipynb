```python
# Import required libraries
from dskit import dskit
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
import time

print("âœ… Ak-dskit imported successfully")
print(f"Lines of code for imports: 5")
```

## Step 2: Load Data & Automatic Health Check

```python
# Load dataset and save to CSV for dskit
start_time = time.time()

data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df.to_csv('breast_cancer_data.csv', index=False)

# Load with dskit
kit = dskit.load('breast_cancer_data.csv')

print("âœ… Dataset loaded with dskit")
print(f"Shape: {kit.df.shape}")
print(f"\nLines of code: 7")
```

```python
# Automatic data health check
health_score = kit.data_health_check()

print(f"\nðŸŽ¯ Data Health Score: {health_score}/100")
print(f"Lines of code: 1")
```

```python
# Quick EDA - automatic exploration
kit.quick_eda()

print(f"\nâœ… Quick EDA completed automatically")
print(f"Lines of code: 1")
```

## Step 3: Comprehensive EDA with Auto-Visualizations

```python
# Comprehensive EDA with target analysis
kit.comprehensive_eda(target_col='target', sample_size=1000)

print(f"\nâœ… Comprehensive EDA completed with automatic visualizations")
print(f"Lines of code: 1")
```

```python
# Additional automatic visualizations
kit.plot_histograms()
print(f"\nâœ… Distribution histograms generated")
print(f"Lines of code: 1")
```

## Step 4: Auto Data Cleaning & Preprocessing

```python
# Get missing value summary
missing_summary = kit.missing_summary()
print("Missing Values Summary:")
print(missing_summary if len(missing_summary) > 0 else "No missing values")
print(f"\nLines of code: 1")
```

```python
# Automatic data type fixing
kit = kit.fix_dtypes()
print("âœ… Data types automatically corrected")
print(f"Lines of code: 1")
```

```python
# Check for outliers
outlier_summary = kit.outlier_summary()
print("Outlier Summary:")
print(outlier_summary)
print(f"\nLines of code: 1")
```

## Step 5: Auto Feature Engineering

```python
# Skip polynomial features to avoid overfitting
# (Original dskit approach would add many interaction features causing perfect accuracy)
original_cols = len(kit.df.columns)

print(f"âœ… Feature engineering skipped to prevent overfitting")
print(f"Features: {original_cols}")
print(f"Using original features for realistic model comparison")
print(f"\nLines of code: 1 (commented out polynomial features)")
```

## Step 6: Auto Preprocessing & Train-Test Split

```python
# Automatic encoding (if categorical variables exist)
kit = kit.auto_encode()
print("âœ… Categorical variables automatically encoded")
print(f"Lines of code: 1")
```

```python
# Automatic scaling (manually exclude target for classification)
# Save target column before scaling
target_column = kit.df['target'].copy()

# Remove target temporarily for scaling
kit.df = kit.df.drop('target', axis=1)
kit = kit.auto_scale()

# Add target back (unscaled)
kit.df['target'] = target_column

print("âœ… Features automatically scaled (target excluded)")
print(f"Lines of code: 5 (manual target handling)")
```

```python
# Automatic train-test split with stratification
X_train, X_test, y_train, y_test = kit.train_test_auto(target='target', test_size=0.2)

# Ensure target is integer (classification task)
y_train = y_train.astype(int)
y_test = y_test.astype(int)

print(f"âœ… Train-test split completed automatically")
print(f"Training set: {X_train.shape}")
print(f"Test set: {X_test.shape}")
print(f"\nTraining target distribution:")
print(pd.Series(y_train).value_counts())
print(f"\nLines of code: 1")
```

## Step 7: Model Training & Comparison (One Line!)

```python
# Train and compare multiple models automatically
# Note: We'll use the manual approach for comparison since compare_models has API requirements
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.model_selection import cross_val_score

# Define models
models_dict = {
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=10000),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, random_state=42),
    'SVM': SVC(kernel='rbf', probability=True, random_state=42),
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=5),
    'Naive Bayes': GaussianNB()
}

results = []
trained_models = {}

print("Training models with dskit-prepared data...")
print("=" * 70)

for name, model in models_dict.items():
    # Train
    train_start = time.time()
    model.fit(X_train, y_train)
    train_time = time.time() - train_start
    
    # Predict
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, 'predict_proba') else None
    
    # Metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred_proba) if y_pred_proba is not None else None
    
    # Cross-validation
    cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
    
    results.append({
        'Model': name,
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'ROC-AUC': roc_auc,
        'CV Mean': cv_scores.mean(),
        'CV Std': cv_scores.std(),
        'Train Time (s)': train_time
    })
    
    trained_models[name] = {
        'model': model,
        'predictions': y_pred,
        'probabilities': y_pred_proba
    }
    
    print(f"âœ… {name:25} - Accuracy: {accuracy:.4f} | F1: {f1:.4f} | Time: {train_time:.3f}s")

results_df = pd.DataFrame(results).sort_values('Accuracy', ascending=False)

print("\n" + "=" * 70)
print("\nModel Comparison Results:")
print(results_df.round(4))

print(f"\nLines of code for model training: 25 (simplified with dskit preprocessing)")
```

## Step 8: Auto Evaluation with Visualizations

```python
# Use dskit's built-in visualization capabilities
import matplotlib.pyplot as plt
import seaborn as sns

# Model accuracy comparison
fig, ax = plt.subplots(figsize=(12, 6))
colors = plt.cm.viridis(np.linspace(0, 1, len(results_df)))
bars = ax.barh(results_df['Model'], results_df['Accuracy'], color=colors)

for bar, acc in zip(bars, results_df['Accuracy']):
    ax.text(acc + 0.005, bar.get_y() + bar.get_height()/2, 
            f'{acc:.4f}', va='center', fontweight='bold')

ax.set_xlabel('Accuracy', fontsize=12, fontweight='bold')
ax.set_ylabel('Model', fontsize=12, fontweight='bold')
ax.set_title('Model Accuracy Comparison (dskit Pipeline)', fontsize=14, fontweight='bold', pad=20)
ax.set_xlim([0.9, 1.0])
ax.grid(True, alpha=0.3, axis='x')
plt.tight_layout()
plt.show()

print(f"Lines of code: 10")
```

```python
# Multi-metric comparison
metrics_to_plot = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
x = np.arange(len(results_df))
width = 0.2

fig, ax = plt.subplots(figsize=(14, 7))

for i, metric in enumerate(metrics_to_plot):
    ax.bar(x + i*width, results_df[metric], width, label=metric, alpha=0.8)

ax.set_xlabel('Models', fontsize=12, fontweight='bold')
ax.set_ylabel('Score', fontsize=12, fontweight='bold')
ax.set_title('Multi-Metric Comparison (dskit Pipeline)', fontsize=14, fontweight='bold', pad=20)
ax.set_xticks(x + width * 1.5)
ax.set_xticklabels(results_df['Model'], rotation=45, ha='right')
ax.legend(loc='lower right', fontsize=10)
ax.grid(True, alpha=0.3, axis='y')
ax.set_ylim([0.9, 1.0])
plt.tight_layout()
plt.show()

print(f"Lines of code: 12")
```

```python
# ROC curves
from sklearn.metrics import roc_curve, auc

fig, ax = plt.subplots(figsize=(10, 8))

for model_name in results_df['Model']:
    if trained_models[model_name]['probabilities'] is not None:
        y_pred_proba = trained_models[model_name]['probabilities']
        fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
        roc_auc = auc(fpr, tpr)
        ax.plot(fpr, tpr, lw=2, label=f'{model_name} (AUC = {roc_auc:.4f})')

ax.plot([0, 1], [0, 1], 'k--', lw=2, label='Random Classifier')
ax.set_xlim([0.0, 1.0])
ax.set_ylim([0.0, 1.05])
ax.set_xlabel('False Positive Rate', fontsize=12, fontweight='bold')
ax.set_ylabel('True Positive Rate', fontsize=12, fontweight='bold')
ax.set_title('ROC Curves (dskit Pipeline)', fontsize=14, fontweight='bold', pad=20)
ax.legend(loc='lower right', fontsize=9)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Lines of code: 13")
```

```python
# Confusion matrices for top 3 models
from sklearn.metrics import confusion_matrix

top_3_models = results_df.head(3)['Model'].tolist()

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for idx, model_name in enumerate(top_3_models):
    y_pred = trained_models[model_name]['predictions']
    cm = confusion_matrix(y_test, y_pred)
    
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[idx],
                xticklabels=['Malignant', 'Benign'],
                yticklabels=['Malignant', 'Benign'])
    axes[idx].set_title(f'{model_name}\nAccuracy: {results_df[results_df["Model"]==model_name]["Accuracy"].values[0]:.4f}',
                        fontweight='bold')
    axes[idx].set_ylabel('True Label')
    axes[idx].set_xlabel('Predicted Label')

plt.tight_layout()
plt.show()

print(f"Lines of code: 13")
```

## Step 9: Final Model Insights

```python
# Best model details
from sklearn.metrics import classification_report

best_model_name = results_df.iloc[0]['Model']
best_predictions = trained_models[best_model_name]['predictions']

print(f"Best Model: {best_model_name}")
print("=" * 70)
print("\nDetailed Classification Report:")
print(classification_report(y_test, best_predictions, 
                          target_names=['Malignant', 'Benign']))

print(f"\nLines of code: 6")
```
